# ðŸ“˜ Linear Regression

### Apa itu Linear Regression?
Linear Regression (Regresi Linier) adalah algoritma supervised learning yang digunakan untuk memprediksi nilai kontinu berdasarkan hubungan linier antara fitur (input) dan target (output).

Contoh Kasus Penggunaan:

*   Memprediksi harga rumah berdasarkan ukuran dan lokasi.
*   Memprediksi hasil ujian berdasarkan waktu belajar.
*   Memprediksi penjualan berdasarkan iklan.

### Konsep Dasar Linear Regression
Model Matematika:

> y=mx+c

*   y = nilai prediksi
*   x = fitur/input
*   m = koefisien/kemiringan garis (slope)
*   c = intercept/konstanta

### Kelebihan & Kekurangan Linear Regression
Kelebihan:
*   Mudah dipahami dan diinterpretasikan.
*   Cepat dan efisien untuk dataset kecil-menengah.
*   Cocok untuk data dengan hubungan linier.

Kekurangan:
*   Tidak cocok untuk hubungan non-linier.
*   Sensitif terhadap outlier.
*   Asumsi multikolinearitas & homoskedastisitas bisa membatasi.



# Contoh Program Python: Linear Regression

```python
# Import library yang dibutuhkan
import numpy as np  # Numpy digunakan untuk operasi numerik seperti array dan perhitungan matematika
import matplotlib.pyplot as plt  # Matplotlib digunakan untuk visualisasi data
from sklearn.linear_model import LinearRegression  # Library machine learning untuk regresi linier
from sklearn.metrics import mean_absolute_percentage_error  # Metrik evaluasi untuk akurasi

# Buat dataset dummy dengan Numpy Array
# Data jumlah kamar dalam rumah
bedrooms = np.array([1, 1, 2, 2, 3, 4, 4, 5, 5, 5]).reshape(-1, 1)  # Perlu reshape untuk sklearn

# Data harga rumah dalam dollar (asumsi)
house_price = np.array([15000, 18000, 27000, 34000, 50000, 68000, 65000, 81000, 85000, 90000])

# Buat model regresi linier
model = LinearRegression()
model.fit(bedrooms, house_price)  # Melatih model dengan data

# Dapatkan koefisien dan intersep
b1 = model.coef_[0]  # Slope atau kemiringan garis regresi
b0 = model.intercept_  # Titik potong dengan sumbu Y

# Prediksi harga rumah berdasarkan model regresi linier
predicted_prices = model.predict(bedrooms)

# Evaluasi model menggunakan akurasi (1 - MAPE)
mape = mean_absolute_percentage_error(house_price, predicted_prices) * 100
accuracy = 100 - mape

print(f'Akurasi Model: {accuracy:.2f}%')

# Buat plot dari model
plt.scatter(bedrooms, house_price, color='blue', label='Data Aktual')  # Plot data asli
plt.plot(bedrooms, predicted_prices, color='red', label='Regresi Linier')  # Garis regresi
plt.xlabel('Jumlah Kamar')  # Label sumbu x
plt.ylabel('Harga Rumah (dalam Dollar)')  # Label sumbu y
plt.legend()  # Menampilkan legenda
plt.title('Regresi Linear: Jumlah Kamar vs Harga Rumah')  # Judul grafik
plt.show()  # Tampilkan plot

# Cetak hasil model regresi
print(f'Persamaan regresi: Harga Rumah = {b0:.2f} + {b1:.2f} * Jumlah Kamar')
```

```python
# Input jumlah kamar dari pengguna
user_bedrooms = int(input("Masukkan jumlah kamar: "))
user_predicted_price = model.predict(np.array([[user_bedrooms]]))[0]
print(f'Prediksi harga rumah untuk {user_bedrooms} kamar: ${user_predicted_price:.2f}')
```
